```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
%matplotlib inline
```

```python
df = pd.read_excel("../data/processed/cleaned_telco_customer_churn.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=df, x="Churn Label")
plt.title("Customer Churn Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=df, x="Gender")
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Contract")
plt.title("Contract Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Internet Service")
plt.title("Internet Service")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Payment Method")
plt.xticks(rotation=30)
plt.title("Payment Method")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Monthly Charges"], bins=30, kde=True)
plt.title("Monthly Charges Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Tenure Months"], bins=30, kde=True)
plt.title("Tenure Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["CLTV"], bins=30, kde=True)
plt.title("Customer Lifetime Value")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Contract", hue="Churn Label")
plt.title("Contract Type vs Churn")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=df, x="Gender", hue="Churn Label")
plt.title("Gender vs Churn")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Internet Service", hue="Churn Label")
plt.title("Internet Service vs Churn")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Payment Method", hue="Churn Label")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Churn Label", y="Monthly Charges")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Churn Label", y="Tenure Months")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(
    df[
        [
            "Monthly Charges",
            "Tenure Months",
            "CLTV",
            "Churn Value"
        ]
    ]
)
plt.show()
```

```python
numeric_df.boxplot(figsize=(15,6))
plt.xticks(rotation=45)
plt.show()
```
