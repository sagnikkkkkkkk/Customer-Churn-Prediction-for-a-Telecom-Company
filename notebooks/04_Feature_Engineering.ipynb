```python
# Customer Churn Prediction for a Telecom Company

# Phase 6 – Feature Engineering

# Objective
# The objective of this notebook is to create new features from the existing data that can improve the predictive performance of the machine learning model.

# Tasks
# - Load the cleaned dataset
# - Create new meaningful features
# - Validate the engineered features
# - Save the engineered dataset
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../data/processed/cleaned_telco_customer_churn.xlsx")
```

```python
df.head()
```

```python
df["Tenure Group"] = pd.cut(
    df["Tenure Months"],
    bins=[0,12,24,48,72],
    labels=["0-12","13-24","25-48","49-72"]
)
```

```python
df["Tenure Group"].value_counts()
```

```python
df["Monthly Charge Group"] = pd.qcut(
    df["Monthly Charges"],
    q=4,
    labels=["Low","Medium","High","Very High"]
)
```

```python
df["Monthly Charge Group"].value_counts()
```

```python
service_columns = [
    "Phone Service",
    "Multiple Lines",
    "Online Security",
    "Online Backup",
    "Device Protection",
    "Tech Support",
    "Streaming TV",
    "Streaming Movies"
]

df["Total Services"] = (
    df[service_columns] == "Yes"
).sum(axis=1)
```

```python
df["Total Services"].describe()
```

```python
df["Senior Citizen"] = df["Senior Citizen"].map(
    {
        "Yes":1,
        "No":0
    }
)
```

```python
df["Partner"] = df["Partner"].map(
    {
        "Yes":1,
        "No":0
    }
)
```

```python
df["Dependents"] = df["Dependents"].map(
    {
        "Yes":1,
        "No":0
    }
)
```

```python
df["Paperless Billing"] = df["Paperless Billing"].map(
    {
        "Yes":1,
        "No":0
    }
)
```

```python
df["Average Monthly Spending"] = (
    df["Total Charges"] /
    df["Tenure Months"].replace(0,1)
)
```

```python
df["Customer Value"] = pd.qcut(
    df["CLTV"],
    q=3,
    labels=["Low","Medium","High"]
)
```

```python
contract_map = {
    "Month-to-month":1,
    "One year":12,
    "Two year":24
}

df["Contract Duration"] = df["Contract"].map(contract_map)
```

```python
df.info()
```

```python
df.to_excel(
    "../data/processed/feature_engineered_telco_customer_churn.xlsx",
    index=False
)
```

```python
# Observations

# The following new features were successfully created:

# - Tenure Group
# - Monthly Charge Group
# - Total Services
# - Average Monthly Spending
# - Customer Value
# - Contract Duration

# These engineered features are expected to improve the predictive capability of the machine learning models.
```
